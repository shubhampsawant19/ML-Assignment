```python
stop_words = set(stopwords.words('english'))
def clean_Data(check):
    check = re.sub(r'http\S+', '', check)
    check = re.sub('[^a-zA-Z]',' ',check)
    check = word_tokenize(check)
    check = [item for item in check if item not in stop_words]
    #check = re.findall('[A-Z][^A-Z]*',check)
    check = ' '.join(check)
    return check
```

```python
data['CleanData'] = data['RAW DATA'].apply(clean_Data)
data.head()
```

```python
def is_capital(data):
    count = 0
    hskills = []
    for kills in data:
        count = 0
        for word in kills.split():
            if word[0].islower():
                count+=1
        if (count>0):
            hskills.append(0)
        else:
            hskills.append(1)
    return hskills
```

```python
req = is_capital(data["RAW DATA"])
```

```python
req_skill = data
req_skill["Flag"] = req
req_skill
```
